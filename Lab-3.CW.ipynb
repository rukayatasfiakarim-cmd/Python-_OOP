```python
#Task-1.ii
class MangoTree:
  def __init__(self, mango_name):
    self.variety= mango_name
    self.height= 1
    self.number_of_mangoes= 0


  def growthUpdate(self, time):
    print(f"Variety: {self.variety}")
    self.height= 1+ 3*time
    print(f"Height: {self.height} meter(s)")
    if self.variety == 'Gopalbhog':
      self.number_of_mangoes= self.height* 10
    elif self.variety == 'Amrapali':
       self.number_of_mangoes= self.height* 15
    print('Number of mangoes on the tree:', self.number_of_mangoes)




print("Updated details after 5 years:")
mangoTree1= MangoTree("Gopalbhog")
print("=====================================")
mangoTree1.growthUpdate(5)
print("=====================================")
mangoTree2= MangoTree("Amrapali")
mangoTree2.growthUpdate(5)
print("=====================================")
```

```python
#Task-2.A
class Customer:
  def __init__(self):
    self.number= 0
    self.amount= 0
    print('Welcome to ABC Memorial Park')


  def buyTicket(self, name, year):
    self.name= name
    self.age= year
    if self.number<3 :
      self.number+= 1

      print(f"Successfully purchased a ticket for {self.name}!")
      if self.age>10:
        self.amount+= 100
      else:
        self.amount+= 50
    else:
      print("You can't buy more than 3 tickets")


  def showDetails(self):
    print(f"Amount of tickets: {self.number}")
    print(f"Total price: {self.amount} Taka")


print('1-------------------------')
customer1 = Customer()
print('2-------------------------')
customer1.buyTicket('Bob', 23)
customer1.buyTicket('Henry', 7)
customer1.buyTicket('Alexa', 30)
customer1.buyTicket('Jonas', 43)
print('3-------------------------')
customer1.showDetails()
print('4-------------------------')
customer2 = Customer()
print('5-------------------------')
customer2.buyTicket('Harry', 60)
customer2.buyTicket('Tomas', 28)
print('6-------------------------')
customer2.showDetails()
```

```python
#Task-3
class GreenPhone:
  def __init__(self, model, version, camera):

    self.model= model
    self.version= version
    self.camera= camera


  def showSpecification(self):
    print('Phone Company: GreenPhone')
    print('Model name:', self.model)
    print('Android Version:', self.version)
    print('Number of Cameras:',self.camera)


  def updatePhone(self):
    self.version+= 1
    if self.model== 'A1' and self.version>14:
      print('Your phone Greenphone A1 is already up to date.')
      self.version= 14
    else:
      print(f"Your phone Greenphone {self.model} is upgraded to Android Version: {self.version}.")

print('1=======================')
p1 = GreenPhone('A1', 12, 3)
p2 = GreenPhone('M11', 12, 4)
p3 = GreenPhone('U20', 12, 5)
p1.showSpecification()
print('2=======================')
p2.showSpecification()
print('3=======================')
p1.updatePhone()
print('4=======================')
p1.updatePhone()
p2.updatePhone()
p3.updatePhone()
print('5=======================')
p1.updatePhone()
p2.updatePhone()
p3.updatePhone()
print('6=======================')
p2.updatePhone()
p3.updatePhone()
print('7=======================')
p1.showSpecification()
p3.showSpecification()
```
