```python
#String.1
given= input()
upcount= 0
lowcount= 0
out= ''
for i in given:
  if 65<=ord(i)<=90:
    upcount+= 1
  elif 97<=ord(i)<=122:
    lowcount+= 1
if upcount>lowcount:
  for i in given:
    if 65<=ord(i)<=90:
      out+= i
    elif 97<=ord(i)<=122:
      out+= chr(ord(i)-32)
elif lowcount>upcount:
  for j in given:
    if 65<=ord(j)<=90:
      out+= chr(ord(j)+32)
    elif 97<=ord(j)<=122:
      out+= j

print(out)
```

```python
#String.2
given= input()
numcount= 0
letcount= 0
for i in given:
  if 48<=ord(i)<=57:
    numcount+=1
  elif 65<=ord(i)<=90 or 97<=ord(i)<=122:
    letcount+=1
if numcount== len(given):
  print('NUMBER')
elif letcount== len(given):
  print('WORD')
else:
  print('MIXED')
```

```python
#List.4

list1= []
while(True):
  i= input()
  if i=='STOP':
   break
  else:
   list1.append(i)


for j in list1:
  list2= j.split(' ')

  list3= []
  for k in list2:
    list3.append(int(k))
  list4= []
  for m in range(len(list3)-1):
    y= abs(list3[m+1]- list3[m])
    list4.append(y)

  flag= True
  for z in list4:
    if z not in range(1, len(list3)):
      flag= False
      break
  if flag== True:
    print('UB Jumper')
  else:
    print('Not UB Jumper')
```

```python
#Dictionary&Tuple.1
inp1= input()
inp2= input()
dict1= {}
dict2= {}
dict3= {}
list1= inp1.split(', ')
list2= inp2.split(', ')
unique= []

for i in list1:
  list3= i.split(': ')
  key= list3[0]
  value= int(list3[1])
  dict1[key]= value

for i in list2:
  list4= i.split(': ')
  key= list4[0]
  value= int(list4[1])
  dict2[key]= value

for k,v in dict1.items():
  if k not in dict3.keys():
    dict3[k]= v
  elif k in dict3.keys():
    dict3[k]+= v


for k,v in dict2.items():
  if k not in dict3.keys():
    dict3[k]= v
  elif k in dict3.keys():
    dict3[k]+= v

for v in dict3.values():
  if v not in unique:
    unique.append(v)


unique.sort()
tup= tuple(unique)

print(dict3)
print(tup)
```

```python
#Dictionary&Tuple.4
s= input()
s= s.upper()
out= ''

dict1= {'1':['.', ',', '?', '!', ':'], '2':['A', 'B', 'C'], '3':['D', 'E', 'F'], '4':['G', 'H', 'I'], '5':['J', 'K', 'L'], '6':['M', 'N', 'O'], '7':['P', 'Q', 'R', 'S'], '8':['T', 'U', 'V'], '9':['W', 'X', 'Y', 'Z'], '0':[' ']}
for i in s:
  for k,v in dict1.items():
    if i in dict1[k]:
      for z in range(len(dict1[k])):
        if dict1[k][z]== i:
          out+=k
          break
        else:
          out+= k

print(out)
```
