```python
#Task-2
class Student:
  def __init__(self, name, id, cgpa):
    self.name= name
    self.__id= id
    self.__cgpa= cgpa


  def setId(self, id):
    self.__id= id

  def getId(self):
    return self.__id

  def getcgpa(self):
    return self.__cgpa




class Department:
  def __init__(self, dept):
    self.dept= dept
    self.idlist= []
    self.stdlist= []

  def findStudent(self, id):
    if id not in self.idlist:
      print(f"Student with this ID doesn't exist, Please give a valid ID")
    else:
      for i in self.stdlist:
        if i.getId()== id:
          print(f"Student info:\nStudent Name: {i.name}\nID: {i.getId()}\nCGPA:  {i.getcgpa()}")


  def addStudent(self,*n):
    for i in n:
      if i.getId() in self.idlist:
        print("Student with the same ID already exists, Please try with another ID")
      else:
        print(f"Welcome to {self.dept} department, {i.name}")
        self.idlist.append(i.getId())
        self.stdlist.append(i)

  def details(self):
    print(f"Department Name: {self.dept}\nNumber of student:{len(self.stdlist)}")
    print(f"Details of the students:")
    for j in self.stdlist:
      print(f"Student name: {j.name}, ID: {j.getId()}, cgpa: {j.getcgpa()}")


s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()



```

```python
#Task-3
class Spaceship:
  def __init__(self, name, capacity):
    self.name= name
    self.capacity= capacity
    self.itemlist= []
    self.weight= 0

  def load_cargo(self, obj1):
    if obj1.getWeight()<=self.capacity:
      self.weight+= obj1.getWeight()
      self.itemlist.append(obj1.getName())
    else:
      print(f"Warning: Unable to load {obj1.getName()} inside {self.name}. Exceeds capacity by {self.weight+obj1.getWeight()-self.capacity}.")


  def display_details(self):
    print(f"Spaceship Name: {self.name}\nCapacity: {self.capacity}\nCurrent Cargo Weight: {self.weight}\nCargo: {self.itemlist}")





class Cargo:
  def __init__(self, name, weight):
    self.__name= name
    self.__weight= weight

  def getName(self):
    return self.__name

  def getWeight(self):
    return self.__weight





# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```

```python
#Task-4

class Student:
  def __init__(self, name, id, dept):
    self.name= name
    self.id= id
    self.dept= dept
    print("Student object is created!")
    self.password= None
    self.email= None
    self.courselist= []


class Usis:
  def __init__(self):
    print("USIS is ready to use!")
    self.advcount= 0

  def login(self, std):
    if std.password== None or std.email== None:
      print("Email and password need to be set.")
    else:
      print("Login sccessful!")

  def advising(self, std, *n):
    if len(n)== 0:
      if self.advcount==0:
        print("Please login to advise courses!")
      else:
        print("You haven't selected any courses.")
    elif len(n)>3:
      print("You need special approval to take more than 3 courses.")
    else:
      for i in n:
        std.courselist.append(i)
      print("Advising successful!")
    self.advcount+=1


  def individualDetails(self, std):
    print(f"Name: {std.name}\nID: {std.id}\nDepartment: {std.dept}")
    s= 'Advised courses: '
    for i in std.courselist:
      s+= i+  ', '
    s= s[:-2:]
    return s


rakib = Student("Rakib", 12301455, "CSE")
print("1***********************")
usis_obj = Usis()
print("2***********************")
usis_obj.login(rakib)
print("3***********************")
usis_obj.advising(rakib)
print("4***********************")
rakib.email = "rakib@hotmail.com"
rakib.password = "1234"
print("5***********************")
usis_obj.login(rakib)
print("6***********************")
usis_obj.advising(rakib)
print("7***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110", "CSE260")
print("8***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110")
print("9***********************")
print(usis_obj.individualDetails(rakib))
```

```python
#Task-5

class Train:
  def __init__(self, name, *n):
    self.name= name
    self.stoplist= []
    self.passlist= []
    for i in n:
      self.stoplist.append(i)
    self.start= self.stoplist[0]
    self.destination= self.stoplist[-1]
    print(f"Welcome aboard on {self.name}\nStart: {self.start}\nDestination: {self.destination}")

  def addPassenger(self, *passenger):
    for i in passenger:
      print(f"{i.name} welcome abroad")
      self.passlist.append(i)

  def allPassengerDetails(self):
    for i in self.passlist:
      if i.start== None:
        i.start= self.start
      if i.destination== None:
        i.destination= self.destination
      count=0
      for j in range (len(self.stoplist)):
        if self.stoplist[j]== i.start:
          for k in range(j+1 ,len(self.stoplist)):
            if self.stoplist[k]== i.destination:
              count+=1
              break
            else:
              count+=1
          break
      i.fair= count*100
      print(f"Name: {i.name},Start: {i.start},Destination: {i.destination},Fair: ${i.fair}")


class Passenger:
  def __init__(self, name, start= None, destination= None):
    self.name= name
    self.start= start
    self.destination= destination

t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print("1========================")
p1 =Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print("5========================")
p4 =Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris")
p6 = Passenger("Mike","Brussels")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()
```

```python
#Task-6

class BracuStudent:
  def __init__(self, name, home):
    self.name= name
    self.home= home
    self.buspass= False


  def show_details(self):
    print(f"Student Name: {self.name}\nLives in {self.home}\nHave Bus Pass? {self.buspass}")


  def get_pass(self):
    self.buspass= True

class BracuBus:
  def __init__(self, route, max= 2):
    self.route= route
    self.max= max
    self.passlist= []


  def board(self, *std):
    if len(std)== 0:
      print("No passengers!")
    else:
      for i in std:
        if len(self.passlist)< self.max:
          if i.buspass== True and i.home== self.route:
            self.passlist.append(i.name)
            print(f"{i.name} boarded the bus.")
          elif i.buspass== False:
            print("You don't have a bus pass!")
          elif i.home!= self.route:
            print("You got on the wrong bus!")
        else:
          print("Bus is full!")

  def show_details(self):
    print(f"Bus Route: {self.route}\nPassengers Count: {len(self.passlist)} (Max: {self.max})\nPassengers On Board: {self.passlist}")


st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()

```
