```python
#Task-4
class abcTech:
  def __init__(self, name, designation, department):
    self.name= name
    self.designation= designation
    self.department= department
    print(f"Welcome to abcTech, {name}!")
    self.skills= []
    self.frame = []

  def addProgrammingSkills(self, list1):
    self.skills+= list1


  def addFrameworks(self, list2):
    self.frame+= list2


  def printInfo(self):
   print(f"Name:  {self.name}")
   print(f"Designation:  {self.designation}")
   print(f"Department: {self.department}")
   string1= ""
   for i in range(len(self.skills)):
    if i== len(self.skills)-1:
      string1+= self.skills[i]
    else:
      string1+= self.skills[i] + ' , '
   print(f"Programming Skills:  {string1}")

   string2= ""
   for j in range(len(self.frame)):
    if j== len(self.frame)-1:
      string2+= self.frame[j]
    else:
      string2+= self.frame[j] + ' , '
   print(f"Frameworks:  {string2}")



  def calculateSalary(self, salary, hour):
    self.hour= hour
    if self.hour>144:
      extra= self.hour - 144
      self.salary= (extra* 800) + salary
    else:
      self.salary= salary
    return self.salary



print("-------------------------")
b1 =abcTech("Tamim Hasan", "Software Engineer", "Android Development")
print("-------------------------")
b1.addProgrammingSkills(["Java", "Python"])
b1.addProgrammingSkills(["Dart", "C++"])
b1.addFrameworks(["Express.js", "React"])
b1.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b1.calculateSalary(45000, 156)}")
print("-------------------------")
print("-------------------------")
b2 =abcTech("Jahin Khandoker", "Senior Developer", "App Development")
print("-------------------------")
b2.addProgrammingSkills(["Java", "Dart", "Swift"])
b2.addFrameworks(["Flutter", "React Native"])
b2.addFrameworks(["Xamarin"])
b2.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b2.calculateSalary(103000, 123)}")
print("-------------------------")
```

```python
#Task-5

class StudentDatabase:
  def __init__(self, name, id):
    self.name= name
    self.id= id
    self.grades={}
    self.credit= 3

  def calculateGPA(self, list1, semester):
    self.semester= semester
    dict1= {}
    list2= []
    sum= 0
    for i in list1:
      x= i.split(': ')
      course= x[0]
      gp= float(x[1])
      list2.append(course)
      sum+= (gp* self.credit)

    cgpa= sum/(len(list1)*self.credit)
    cgpa= round(cgpa,2)

    tup1= tuple(list2)
    dict1[tup1]= cgpa

    self.grades[self.semester]= dict1

  def printDetails(self):

    print(f"Name: {self.name}")
    print(f"ID: {self.id}")
    for k,v in self.grades.items():
      print(f"Courses taken in {k}:")
      for k2,v2 in v.items():
        for i in k2:
          print(i)

        print(f"GPA: {v2}")

s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()

```

```python
#Task-6


class Store:
  def __init__(self, balance):
    print('New branch created!')
    self.balance= balance
    self.total_items= 0
    self.dict1= {}


  def viewAllItems(self):
    if self.total_items== 0:
      print('There is no items in your inventory')
    else:
      string1= 'All items: '
      for k,v in self.dict1.items():
        string1+= k + ', '

      string1= string1.strip(', ')

      print(string1)



  def viewAllItemDetails(self):
      print(self.dict1)



  def add_item(self, item_list):
    item_name= item_list[0]
    print(f"Item added: {item_name}")
    self.total_items +=1

    stock= item_list[1]
    buying_price= item_list[2]
    selling_price= item_list[3]
    self.balance-= buying_price *stock

    self.dict1[item_name]= {'stock': stock, 'buying_price': buying_price, 'selling_price': selling_price}



  def sell_item(self, item_name, quantity):
    if self.dict1[item_name]['stock'] >= quantity:
      self.dict1[item_name]['stock']-= quantity
      print(self.dict1)
      self.balance+= (self.dict1[item_name]['selling_price']) *quantity

    else:
      print(f"Sorry! {item_name} is not available at your desired quantity. Currently we have: {self.dict1[item_name]['stock']}")


  def restock_item(self, item_name, quantity):
    self.dict1[item_name]['stock'] += quantity
    print(f"Restocked item: {item_name},")
    print(f"Current Stock: {self.dict1[item_name]['stock']}")

    self.balance-= (self.dict1[item_name]['buying_price']) *quantity


print("==========================")
branch1 = Store(5000)
print(f"Current Balance: {branch1.balance}")
print(f"Total items: {branch1.total_items}")
branch1.viewAllItems()
branch1.viewAllItemDetails()
print("==========================")
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["ChaCha Noodles", 10, 5, 8])
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["Sparrow Shampoo", 5, 10, 20])
print(f"Current Balance: {branch1.balance}")
print("==========================")
branch1.viewAllItems()
print()
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.sell_item("ChaCha Noodles", 15)
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
branch1.sell_item("ChaCha Noodles", 10)
print()
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.restock_item("ChaCha Noodles", 5)
print()
branch1.viewAllItemDetails()
print()
print(f"Current Balance: {branch1.balance}\n")
print("==========================")




```
