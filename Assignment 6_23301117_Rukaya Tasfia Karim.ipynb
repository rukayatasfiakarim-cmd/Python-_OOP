```python
#Task-1
class Student:
  id= 0
  def __init__(self, name, dept, age, cgpa):
    self.name= name
    self.dept= dept
    self.age= age
    self.cgpa= cgpa
    Student.id+= 1


  def showDetails(self):
    print(f"ID: {Student.id}\nName: {self.name}\nDepartment: {self.dept}\nAge: {self.age}\nCGPA: {self.cgpa}")


  @classmethod
  def from_String(cls, info):
    name, dept, age, cgpa= info.split("-")
    obj= cls(name, dept, int(age), float(cgpa))
    return obj


s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails()

#Subtask-5
print()
print("Class variables are defined within a class but outside of any class methods.\nInstance variables are defined within class methods, typically inside the constructor.")
print("Changing the class variable value will also change all the instance value that inherited the same variable from the class,\nBut changing an instance variable will only change the varialbe value inside of that specific instance.\n\n\n")

#Subtask-6
print("Instance methods stay within a class and they need a class instance and can access the instance through self typically.\nClass methods don't need a class instance.\nThey have access to their class itself to change or manipulate something globally or before sending it to the constructor.")

```

```python
#Task-2
class Passenger:
  count= 0
  def __init__(self, name):
    self.name= name
    Passenger.count+=1


  def set_bag_weight(self, weight):
    if weight<=20:
      self.fare= 450
    elif 20<weight<=50:
      self.fare= 500
    elif weight>50:
      self.fare= 550

  def printDetail(self):
    print(f"Name: {self.name}\nBus Fare: {self.fare} taka")




print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)

```

```python
#Task-4
class NikeBangladesh:
  branchlist= []
  itemdict= {'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe': 0}
  sold= 0
  def __init__(self, branch):
    self.outlet= branch
    if branch not in NikeBangladesh.branchlist:
      NikeBangladesh.branchlist.append(branch)
    self.sold= 0
    self.itemdict= {'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe': 0}

  @classmethod
  def status(cls):
    print(f"Nike Bangladesh Status:\nBranches Opened: {cls.branchlist}\nCurrently Stocked\n{cls.itemdict}\nSold: {cls.sold}")

  def details(self):
    print(f"Nike {self.outlet} outlet:\nProducts Currently Stocked: \n{self.itemdict}\nSold: {self.sold}")

  def restockProducts(self, dict1):
    for k, v in dict1.items():
      self.itemdict[k]+= v
      NikeBangladesh.itemdict[k]+= v


  def productSold(self, dict2):
    for k, v in dict2.items():
      self.itemdict[k]-= v
      NikeBangladesh.itemdict[k]-= v
      self.sold+= v
      NikeBangladesh.sold+= v

print("xxxxxxxxxxxxxx1xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
dhaka = NikeBangladesh("Dhaka Banani")
chittagong = NikeBangladesh("Chittagong GEC")
print("xxxxxxxxxxxxxx2xxxxxxxxxxxxxxxx")
dhaka.details()
print("xxxxxxxxxxxxxx3xxxxxxxxxxxxxxxx")
chittagong.details()
print("xxxxxxxxxxxxxx4xxxxxxxxxxxxxxxx")
dhaka.restockProducts(
{"Air Jordan":1200,"Cortez":200,"Zoom Kobe":200})
chittagong.restockProducts(
{"Air Jordan":1000,"Cortez":250,"Zoom Kobe":100})
print("xxxxxxxxxxxxxx5xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
print("xxxxxxxxxxxxxx6xxxxxxxxxxxxxxxx")
dhaka.productSold({"Air Jordan":760,"Cortez":90})
chittagong.productSold({"Air Jordan":520,"Zoom Kobe":70})
print("xxxxxxxxxxxxxx7xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
```
