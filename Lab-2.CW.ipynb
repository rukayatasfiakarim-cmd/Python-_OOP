```python
#Task-1
def BMI(height, weight):
  h= height/100
  w= weight
  score= w/(h**2)
  score= round(score,1)
  return score

score= BMI(175, 96)
if score<18.5:
  decision= "Underweight"
elif 18.5<=score<=24.9:
  decision= "Normal"
elif 25<=score<=30:
  decision= 'Overweight'
elif score>30:
  decision= 'Obese'
print(f"Score is {score}. You are {decision}.")
```

```python
#Task-2
dict1= {'BBQ Chicken Cheese Burger':250, 'Beef Burger':170, 'Naga Drums': 200}
def chillox(order, place= "Mohakhali"):
  for k,v in dict1.items():
    if order==k:
      meal= v
  if place== 'Mohakhali':
    delivery= 40
  else:
    delivery= 60
  tax= meal*8/100
  cost= meal+ delivery+ tax
  return cost


cost= chillox('Beef Burger', 'Dhanmondi')
print(cost)
```

```python
#Task-3
def email(address, new, old= 'sheba.xyz'):
  if new==old:
    decision= 'Unchanged'
  else:
    decision= 'Changed'
  list1= address.split('@')
  result= list1[0]+'@'+new
  return decision,result
decision, result= email('alice@kaaj.com', 'sheba.xyz', 'kaaj.com')
print(f"{decision}: {result}")
```

```python
#Task-4
def palindrome(s):
  new= ''
  count= 0
  minus=-1
  for i in s:
    if i==' ':
      pass
    else:
      new+=i

  middle= (len(new)//2)

  for i in range(middle):
    if new[i]==new[minus]:
      count+=1
    minus-=1
  if count== middle:
    print('Palindrome')
  else:
    print('Not a palindrome')


palindrome('madam')
```

```python
#Task-5
def time(day):
  year= day//365
  day= day%365
  month= day//30
  day= day%30
  return year, month, day
year, month,day=time (4320)
print(f"{year} years, {month} months and {day} days")
```
