```python
#Task-1
def hospital_fee(**kwargs):
  max_amount= 0
  max_payer= []
  for key,value in kwargs.items():
    if value>max_amount:
      max_amount= value

  for key,value in kwargs.items():
    if value== max_amount:
      max_payer.append(key)


  return max_amount, max_payer


max_amount, max_payer = hospital_fee(Neymar =
1000, Dembele = 600, Reus = 500, Bale = 1000)

string1= str(max_payer)[1:-1:]
string2= ''
for j in string1:
  if j!= "'":
    string2+= j

print(f"Highest fee was {max_amount} tk which was paid by {string2}.")
```

```python
#Task-2
def is_james_bond(list1):
  s= ''
  for i in list1:
    if i== 0 or i== 7:
      s+= str(i)
  return s

s= is_james_bond([1, 0, 2, 0, 4, 7, 5] )
if '007' in s:
  print('True')
else:
  print('False')
```

```python
#Task-3
def assign_students_to_sections(sec, *name):
  dict1= {}
  dict2={}
  for i in name:
    sum= 0
    list1= []
    for j in range(len(i)):
      sum+= ord(i[j])
    snum= sum%len(sec)
    list1.append(i)
    if sec[snum] not in dict1.keys():
      dict1[sec[snum]]= list1
    else:
      dict1[sec[snum]]+= list1
  for z in range(len(sec)):
    if sec[z] in dict1.keys():
      for key, value in dict1.items():
        if sec[z]== key:
          dict2[key]= value
    else:
      dict2[sec[z]]= []

  return dict2

dict2= assign_students_to_sections ('ABCDE', 'Alice',
'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace')
print(dict2)
```

```python
#Task-4
def username_generator(first_name, last_name, student_id, middle_name=''):
  username= ''
  if len(first_name)>2:
    for i in range(0,3):
      if 97<=ord(first_name[i])<=122:
        username+= chr((ord(first_name[i]))-32)
      else:
        username+= first_name[i]
  else:
    for i in range(len(first_name)):
      if 97<=ord(first_name[i])<=122:
        username+= chr((ord(first_name[i]))-32)
      else:
        username+= first_name[i]


  if len(middle_name)>0:
    username+= middle_name

  for j in range(-3,0,1):
    if 65<=ord(last_name[j])<=90:
      username+= chr(ord(last_name[j])+32)
    else:
      username+= last_name[j]
  username+='_'
  username+= str(student_id)[-4::1]

  return username

first_name, middle_name, last_name, student_id= input ("First Name:"), input("Middle Name:"), input ("Last Name:"), int (input ("Student ID:"))

if middle_name== '':
  print(username_generator (first_name, last_name, student_id))
else:
  print(username_generator(first_name, last_name, student_id, middle_name))


```

```python
#Task-5
def key_generator(*names):
  list1= list(names)
  key_list= []
  for i in list1:
    s= ''
    if ord(i[0]) in range(65,91):
      s+= chr(ord(i[0])+32)
    else:
      s+= i[0]


    for j in range(len(i)-2,0,-1):
      s+= str(ord(i[j]))



    if ord(i[-1]) in range(97,123):
      s+= chr(ord(i[-1])-32)
    else:
      s+= i[-1]

    key_list.append(s)
  return key_list



key_list = key_generator ("Alex",
"Bob", "Trudy")

print('Encrypted keys:', key_list)
```

```python
#Task-6
def playRockPaperScissor(rounds):
  import random
  user= 0
  computer= 0
  for i in range(rounds):
    inp= input()
    x= random.choice(['rock', 'paper', 'scissor'])
    print('Computer:',x)
    if inp== 'rock':
      if x== 'paper':
        computer+=1
      elif x== 'scissor':
        user+=1
    elif inp== 'paper':
      if x== 'scissor':
        computer+= 1
      elif x== 'rock':
        user+= 1
    elif inp== 'scissor':
      if x== 'rock':
        computer+= 1
      elif x== 'paper':
        user+= 1

  if computer>user:
    winner= 'computer'
  elif user> computer:
    winner= 'user'
  elif user==computer:
    winner= 'Tie'
  return user, computer, winner

user, computer, winner= playRockPaperScissor(int(input()))
print('Your score:',user)
print("Computer's score:",computer)


if winner== 'computer':
  print('Computer has won the game!')
elif winner== 'user':
  print('You have won the game!')
elif winner== 'Tie':
  print("It's a tie!!!")

```
